```python
df = df.rename(columns={" Mean of the integrated profile" : "ip_Mean",
                        " Standard deviation of the integrated profile" : "ip_Standard_deviation",
                        " Excess kurtosis of the integrated profile": "ip_kurtosis",
                        " Skewness of the integrated profile": "ip_Skewness",
                        " Mean of the DM-SNR curve": "Mean_DM-SNR",
                        " Standard deviation of the DM-SNR curve": "DM-SNR_Standard_deviation",
                        " Excess kurtosis of the DM-SNR curve": "DM-SNR_kurtosis",
                        " Skewness of the DM-SNR curve": "DM-SNR_Skewness",
                        "target_class": "target_class"})
```

```python
df.head()
```

```python
x = df.drop(columns=['target_class'])
y = df['target_class']
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train.shape, x_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_classifier = RandomForestClassifier(n_estimators=100, n_jobs=-1, max_depth=5, oob_score=True, random_state=42)
rf_classifier.fit(x_train, y_train)
```

```python
y_pred = rf_classifier.predict(x_test)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
svc=SVC()
svc.fit(x_train,y_train)
y_pred=svc.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
svc=SVC(C=100.0)
svc.fit(x_train,y_train)
y_pred=svc.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
svc=SVC(C=1000.0)
svc.fit(x_train,y_train)
y_pred=svc.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
liner_svc=SVC(kernel='linear',C=1.0)
liner_svc.fit(x_train,y_train)
y_pred=liner_svc.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
liner_svc=SVC(kernel='linear',C=100.0)
liner_svc.fit(x_train,y_train)
y_pred=liner_svc.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
liner_svc=SVC(kernel='linear',C=1000.0)
liner_svc.fit(x_train,y_train)
y_pred=liner_svc.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
liner_svc=SVC(kernel='poly',C=1.0)
liner_svc.fit(x_train,y_train)
y_pred=liner_svc.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
liner_svc=SVC(kernel='sigmoid',C=1.0)
liner_svc.fit(x_train,y_train)
y_pred=liner_svc.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
liner_svc=SVC(kernel='sigmoid',C=100.0)
liner_svc.fit(x_train,y_train)
y_pred=liner_svc.predict(x_test)
accuracy_score(y_test,y_pred)
```
